```python
from matplotlib import pyplot as plt
from PIL import Image
import cv2
```

```python
bonafide = "/mnt/nas3/yrkim/liveness_lidar_project/GC_project/LDFAS/1. Indoor/3/bonafide/rgb_1.jpg"
print_attack = "/mnt/nas3/yrkim/liveness_lidar_project/GC_project/LDFAS/1. Indoor/3/attack_paper/rgb_1.jpg"
replay_attack = "/mnt/nas3/yrkim/liveness_lidar_project/GC_project/LDFAS/1. Indoor/3/attack_replay/rgb_1.jpg"
mask_attack = "/mnt/nas3/yrkim/liveness_lidar_project/GC_project/LDFAS/1. Indoor/3/attack_mask/rgb_1.jpg"

bonafide_pc = "/mnt/nas3/yrkim/liveness_lidar_project/GC_project/LDFAS/1. Indoor/3/bonafide/pc_1.jpg"
print_attack_pc = "/mnt/nas3/yrkim/liveness_lidar_project/GC_project/LDFAS/1. Indoor/3/attack_paper/rgb_1.jpg"

bona_bgr = cv2.imread(bonafide)
pa_bgr = cv2.imread(print_attack)
ra_bgr = cv2.imread(replay_attack)
ma_bgr = cv2.imread(mask_attack)

bona = cv2.cvtColor(bona_bgr, cv2.COLOR_BGR2RGB)
pa = cv2.cvtColor(pa_bgr, cv2.COLOR_BGR2RGB)
ra = cv2.cvtColor(ra_bgr, cv2.COLOR_BGR2RGB)
ma = cv2.cvtColor(ma_bgr, cv2.COLOR_BGR2RGB)


plt.imshow(bona)
plt.show()

plt.imshow(pa)
plt.show()
```

```python
plt.hist(bona.ravel(), 256, [0,256])
plt.show()

plt.hist(pa.ravel(), 256, [0,256])
plt.show()
```

```python
color = ('r','g','b')
for i, col in enumerate(color):
    histr = cv2.calcHist([bona], [i], None, [256],[0,256])
    plt.plot(histr, color = col)
    plt.xlim([0,256])
plt.title("bonafide")
plt.show()
```

```python
color = ('r','g','b')
for i, col in enumerate(color):
    histr = cv2.calcHist([pa], [i], None, [256],[0,256])
    plt.plot(histr, color = col)
    plt.xlim([0,256])
plt.title("print attack")
plt.show()
```
