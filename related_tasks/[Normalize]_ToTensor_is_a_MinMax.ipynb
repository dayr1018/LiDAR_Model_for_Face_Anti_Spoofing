```python
from PIL import Image
import numpy as np
from torchvision import transforms 
import torch
```

# 원본 (numpy)

```python
rgb_path = "rgb.jpg"
rgb = Image.open(rgb_path)
rgb_np = np.array(rgb)
print(rgb_np.shape)
rgb_np
```

# MinMax 정규화

```python
rgb_minmax = rgb_np / 255
print(rgb_minmax.shape)
rgb_minmax
```

# MinMax한 numpy에 ToTensor 수행 그리고 다시 numpy로 변경 (동일한 값 기대함)

```python
transformer = transforms.Compose([transforms.ToTensor()])

# tensor = torch.from_numpy(rgb_minmax)
tensor = transformer(rgb_minmax)
print(tensor.shape)
tensor
```

```python
back_np = tensor.numpy()
print(back_np.shape)
back_np
```

# 원본(numpy)를 ToTensor해서 변경 (MinMax했던 결과와 동일하길 기대)

```python
transformer = transforms.Compose([transforms.ToTensor()])

# tensor = torch.from_numpy(rgb_minmax)
tensor2 = transformer(rgb_np)
print(tensor2.shape)
tensor2
```
