```python
import pandas as pd
import numpy as np
from plyfile import PlyData
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
plydata = PlyData.read('pointcloud_janghoon.ply')
ply_pd = pd.DataFrame({key_: plydata['vertex'][key_] for key_ in ['x', 'y', 'z', 'red', 'green', 'blue', 'cx', 'cy', 'depth', 'alpha']})
ply_pd
```

```python
ply_pd.max()
```

```python
ply_pd.min()
```

# MinMax 스케일러

```python
minMaxScaler = MinMaxScaler()
# minMaxScaler.fit(ply_pd)
# minmax_pd = minMaxScaler.transform(ply_pd)
minmax_result = minMaxScaler.fit_transform(ply_pd)
minmax_pd = pd.DataFrame(minmax_result)
minmax_pd.columns = ["x","y","z","red","blue","green","cx","cy","depth","alpha"]
minmax_pd
```

```python
minmax_pd[['x']]
```

```python
x_pd = minmax_pd[['x']].copy()
x_numpy = x_pd.to_numpy()
x_numpy
```

# 표준화 스케일러

```python
# standardScaler = StandardScaler(with_mean=False, with_std=False)
standardScaler = StandardScaler()
# ply_pd 는 위에서 불러옴 
standard_result = standardScaler.fit_transform(ply_pd)
standard_pd = pd.DataFrame(standard_result)
standard_pd.columns = ["x","y","z","red","blue","green","cx","cy","depth","alpha"]
standard_pd
```
