```python
# xcoor = []
# for i in range(1,10):
#     ply = PlyData.read(f"/mnt/nas3/yrkim/liveness_lidar_project/GC_project/LDFAS/1. Indoor/{i}/bonafide/pc_{index}.ply")
#     ply_pd = pd.DataFrame({key:ply['vertex'][key] for key in ['x', 'y', 'z', 'red', 'green', 'blue', 'cx', 'cy', 'depth', 'alpha']})
#     xcoor.append(ply_pd['x'])

# for i in range(9):    
#     xcoor[i] = (xcoor[i] - xcoor[i].min()) / (xcoor[i].max() - xcoor[i].min())
#     xcoor[i] = 255*xcoor[i]


cloud_path = ["/mnt/nas3/yrkim/liveness_lidar_project/GC_project/LDFAS/1. Indoor/1/bonafide/pc_1.ply",
           "/mnt/nas3/yrkim/liveness_lidar_project/GC_project/LDFAS/2. Outdoor/5/bonafide/pc_1.ply",
           "/mnt/nas3/yrkim/liveness_lidar_project/GC_project/LDFAS/3. Indoor_dark/7/bonafide/pc_1.ply"]    
xcoor, ycoor, zcoor = [], [], []
for path in cloud_path:
    ply = PlyData.read(path)
    ply_pd = pd.DataFrame({key:ply['vertex'][key] for key in ['x', 'y', 'z', 'red', 'green', 'blue', 'cx', 'cy', 'depth', 'alpha']})
    xcoor.append(ply_pd['x'])
    ycoor.append(ply_pd['y'])
    zcoor.append(ply_pd['z'])
    
    
for i in range(3):    
    xcoor[i] = (xcoor[i] - xcoor[i].min()) / (xcoor[i].max() - xcoor[i].min())
    xcoor[i] = 255*xcoor[i]
    ycoor[i] = (ycoor[i] - ycoor[i].min()) / (ycoor[i].max() - ycoor[i].min())
    ycoor[i] = 255*ycoor[i]
    zcoor[i] = (zcoor[i] - zcoor[i].min()) / (zcoor[i].max() - zcoor[i].min())
    zcoor[i] = 255*zcoor[i]

for i in range(3):    
    plt.hist(xcoor[i], 256, [0,256], color='purple')
    plt.show()    
    
# plt.figure(figsize=(15,10))
# for i in range(3):
#     plt.subplot(3,3,i+1)
#     plt.hist(xcoor[i], 256, [0,256], color='purple')
# plt.show()
```
