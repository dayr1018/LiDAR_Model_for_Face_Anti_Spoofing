```python
# directory = "0810_paper_rgb"
directory = "0810_paper_rgbp"
# directory = "0810_paper_rgbd"
# directory = "0810_paper_rgbdp"

# 첫번째 수정
# directory = "0811_paper_rgb"
# directory = "0811_paper_rgbp"

# 두번째 수정 (스케줄러 추가한 것)
# directory = "0812_paper_rgb"
# directory = "0812_paper_rgbp"


# directory = "0810_replay_rgb"
# directory = "0810_replay_rgbp"
# directory = "0810_replay_rgbd"
# directory = "0810_replay_rgbdp"

# directory = "0810_mask_rgb"
# directory = "0810_mask_rgbp"
# directory = "0810_mask_rgbd"
# directory = "0810_mask_rgbdp"

# directory = "0810_total_rgb"
# directory = "0810_total_rgbp"
# directory = "0810_total_rgbd"
# directory = "0810_total_rgbdp"


indoor = "/mnt/nas3/yrkim/liveness_lidar_project/GC_project/bc_output/logs/" + directory + "/" + directory + "_indoor.pkl"
outdoor = "/mnt/nas3/yrkim/liveness_lidar_project/GC_project/bc_output/logs/" + directory + "/" + directory + "_outdoor.pkl"
dark = "/mnt/nas3/yrkim/liveness_lidar_project/GC_project/bc_output/logs/" + directory + "/" + directory + "_dark.pkl"

with open(indoor,"rb") as fr:
    indoor_check = pickle.load(fr)
with open(outdoor,"rb") as fr:
    outdoor_check = pickle.load(fr)
with open(dark,"rb") as fr:
    dark_check = pickle.load(fr)
```

```python
indoor_check.keys()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(indoor_check['train_loss'])
plt.title('train_loss')
plt.subplot(1,2,2)
plt.plot(indoor_check['test_loss'],color='red')
plt.title('test_loss')
plt.show()


plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(indoor_check['train_f1'])
plt.title('train_f1')
plt.subplot(1,2,2)
plt.plot(indoor_check['test_f1'],color='red')
plt.title('test_f1')
plt.show()

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(indoor_check['train_acc'])
plt.title('train_acc')
plt.subplot(1,2,2)
plt.plot(indoor_check['test_acc'],color='red')
plt.title('test_acc')
plt.show()

# train_epoch = indoor_check['train_f1'].index(max(indoor_check['train_f1']))
# test_epoch = indoor_check['test_f1'].index(max(indoor_check['test_f1']))

# print(f"\tIndoor(train)'s max f1 score\t\t\t\t\tIndoor(test)_door's max f1 score") 
# print(f"\tepoch: {train_epoch}\t\t\t\t\t\t\tepoch: {test_epoch}")
# print(f"\tf1(indoor): {indoor_check['test_f1'][train_epoch]}\t\t\t\t\tf1(indoor): {indoor_check['test_f1'][test_epoch]}") 
# print(f"\tf1(outdoor): {outdoor_check['test_f1'][train_epoch]}\t\t\t\t\tf1(outdoor): {outdoor_check['test_f1'][test_epoch]}")
# print(f"\tf1(dark): {dark_check['test_f1'][train_epoch]}\t\t\t\t\tf1(dark): {dark_check['test_f1'][test_epoch]}")

```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(outdoor_check['test_loss'],color='green')
plt.title('outdoor_loss')
plt.subplot(1,2,2)
plt.plot(dark_check['test_loss'],color='black')
plt.title('dark_loss')
plt.show()

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(outdoor_check['test_f1'],color='green')
plt.title('test_f1')
plt.subplot(1,2,2)
plt.plot(dark_check['test_f1'],color='black')
plt.title('test_f1')
plt.show()

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(outdoor_check['test_acc'],color='green')
plt.title('test_acc')
plt.subplot(1,2,2)
plt.plot(dark_check['test_acc'],color='black')
plt.title('test_acc')
plt.show()


out_epoch = outdoor_check['test_f1'].index(max(outdoor_check['test_f1']))
dark_epoch = dark_check['test_f1'].index(max(dark_check['test_f1']))

print(f"\tOutdoor's max f1 score\t\t\t\t\t\tDark_door's max f1 score") 
print(f"\tepoch: {out_epoch}\t\t\t\t\t\t\tepoch: {dark_epoch}")
print(f"\tacc(indoor): {indoor_check['test_acc'][out_epoch]}\t\t\t\t\t\tacc(indoor): {indoor_check['test_acc'][dark_epoch]}") 
print(f"\tf1(indoor): {indoor_check['test_f1'][out_epoch]}\t\t\t\t\tf1(indoor): {indoor_check['test_f1'][dark_epoch]}") 
print(f"\tacc(outdoor): {outdoor_check['test_acc'][out_epoch]}\t\t\t\tacc(outdoor): {outdoor_check['test_acc'][dark_epoch]}")
print(f"\tf1(outdoor): {outdoor_check['test_f1'][out_epoch]}\t\t\t\t\tf1(outdoor): {outdoor_check['test_f1'][dark_epoch]}")
print(f"\tacc(dark): {dark_check['test_acc'][out_epoch]}\t\t\t\t\tacc(dark): {dark_check['test_acc'][dark_epoch]}")
print(f"\tf1(dark): {dark_check['test_f1'][out_epoch]}\t\t\t\t\tf1(dark): {dark_check['test_f1'][dark_epoch]}")
```
