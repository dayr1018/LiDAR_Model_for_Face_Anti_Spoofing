```python
import os
import os.path as osp
import time
from pathlib import Path
```

```python
# dir_path = osp.join(os.getcwd(), "Indoor_dark_replay")
# dir_path = osp.join(os.getcwd(), "Outdoor_replay")
dir_path = osp.join(os.getcwd(), "test")

# new_path = osp.join(os.getcwd(), "soreted_outdoor")
new_path = osp.join(os.getcwd(), "output_files")
# os.mkdir(new_path)
for i in range(1,13):
    os.makedirs(osp.join(new_path, str(i)))

dir_path
```

```python
files = os.listdir(dir_path)
files
len(files)
```

```python
for file in files:
    file_dir = osp.join(dir_path, file)
    print(f"{file} : {time.ctime(osp.getmtime(file_dir))} ")
```

```python
sorted_files = sorted(Path(dir_path).iterdir(), key=os.path.getmtime, reverse=False)
sorted_files
```

```python
sorted_files[0].name
```

```python
rgb, depth, pc = 1, 1, 1
index = 1
for idx, file in enumerate(sorted_files, start=1):
    
    origin_name = osp.join(dir_path, file.name)
    if "RGB" in file.name:
        filename = "rgb_" + str(rgb) + ".jpg"
        rgb += 1 
    if "Depth" in file.name:
        filename = "depth_" + str(depth) + ".jpg"
        depth += 1 
    if "ply" in file.name:
        filename = "pc_" + str(pc) + ".ply"
        pc += 1 
    new_name = osp.join(new_path, str(index), filename)
    os.rename(origin_name, new_name)
    
    if rgb==21 and depth==21 and pc==21:
        rgb = 1
        depth = 1
        pc = 1
        index += 1
        print(f"index {index}")
```
